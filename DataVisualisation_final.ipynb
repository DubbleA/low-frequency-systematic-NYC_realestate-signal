```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt



# Read the NYC zip code shapefile (download from https://data.cityofnewyork.us/download/i8iw-xf4u/application%2Fzip)
#nyc_zip_shapefile = 'ZIP_CODE_040114.shp'
df = pd.read_csv("MovAvgRest-3.csv", index_col=0)
#df = pd.read_csv("TreeScalFac.csv", index_col=0)

# Read the NYC zip code shapefile (download from https://data.cityofnewyork.us/download/i8iw-xf4u/application%2Fzip)
nyc_zip_shapefile = 'ZIP_CODE_040114.shp'
gdf = gpd.read_file(nyc_zip_shapefile)

# Convert the CSV data to a dictionary of zip codes and scores
zip_scores = df.to_dict('records')[0]

# Merge the zip code scores with the geodataframe
gdf['ZIPCODE'] = gdf['ZIPCODE']

gdf['score'] = gdf['ZIPCODE'].map(zip_scores) * -1 + 25

# Plot the NYC zip codes with a continuous colormap
ax = gdf.plot(column='score', cmap='coolwarm', figsize=(15, 15), linewidth=0.5, edgecolor='black', legend=True, missing_kwds={'color': 'gray'})
ax.set_axis_off()
#plt.title("NYC Zip Codes Colored by Restaurant Inpection Score")
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the CSV file with tree scores
df = pd.read_csv("TreeScalFac.csv", index_col=0)

# Read the NYC zip code shapefile (download from https://data.cityofnewyork.us/download/i8iw-xf4u/application%2Fzip)
nyc_zip_shapefile = 'ZIP_CODE_040114.shp'
gdf = gpd.read_file(nyc_zip_shapefile)

# Define the selected and in-between zip codes
selected_zips = ['10011', '10016', '10023', '10001', '11201', '11530']
inbetween_zips = ['10002', '10003', '10004', '10005', '10006', '10007', '10009', '10010', '10012', '10013', '10014', '10017', '10018', '10019', '10020', '10021', '10022', '10024', '10025', '10027', '10028', '10029', '10030', '10031', '10032', '10033', '10034', '10035', '10036', '10037', '10038', '10039', '10040', '10044', '10065', '10075', '10128', '11205', '11217', '11238', '11542', '11545', '11576']
gdf['selected'] = gdf['ZIPCODE'].isin(selected_zips)
gdf['inbetween'] = gdf['ZIPCODE'].isin(inbetween_zips) & ~gdf['selected']

# Convert the CSV data to a dictionary of zip codes and scores
zip_scores = df.to_dict('records')[0]

# Merge the zip code scores with the geodataframe
gdf['score'] = gdf['ZIPCODE'].map(zip_scores)

# Plot the selected and in-between zip codes with different colors
selected_gdf = gdf.loc[gdf['selected']]
inbetween_gdf = gdf.loc[gdf['inbetween']]
ax = selected_gdf.plot(color='green', figsize=(15, 15), linewidth=0.5, edgecolor='black', legend=True, missing_kwds={'color': 'gray'})
inbetween_gdf.plot(ax=ax, color='gray', linewidth=0.5, edgecolor='black')
ax.set_axis_off()
#ax.set_title("Major REIT Investments")
# ax.set_xlim([-74.03, -73.92])
# ax.set_ylim([40.7, 40.88])
plt.show()
```
